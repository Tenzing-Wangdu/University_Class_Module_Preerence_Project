```python
CIS9665 = pd.read_csv("/Users/wangdu/Desktop/Remote Studies Affects_March 20, 2022_09.16.csv")
CIS9665.head()
```

```python
data = {'In Person':7, 'Fully Remote':9, 'Hybrid':22}
Class_Preference = list(data.keys())
Number_of_respondants = list(data.values())
  
fig = plt.figure(figsize = (6, 5))
 
plt.barh(Class_Preference, Number_of_respondants, color ='b', alpha= 0.3)
 
plt.ylabel("Class preference", size=12)
plt.xlabel("Number of respondants", size =12)
plt.title("Class Preference of CIS 9655 students")
plt.xticks(range(0,26,5))
plt.annotate(xy = [22, 2], text = '22', size=12)
plt.annotate(xy = [9, 1], text = '9', size=12)
plt.annotate(xy = [7, 0], text = '7', size=12)
plt.show()
fig.savefig("CP for CLass.png")
```

```python
data = {'In Person':3, 'Remote':34}
Exams_Preference = list(data.keys())
Number_of_respondants = list(data.values())
  
fig = plt.figure(figsize = (6, 5))
 
plt.barh(Exams_Preference, Number_of_respondants, color ='g', alpha= 0.3)
 
plt.ylabel("Exam preference", size=12)
plt.xlabel("Number of respondants", size =12)
plt.title("Exam Preference of CIS 9655 students")
plt.xticks(range(0,45,5))
plt.annotate(xy = [34, 1], text = '92%', size=12)
plt.annotate(xy = [3, 0], text = '8%', size=12)
fig.savefig("EP for CLass.png")
plt.show()
```

```python
data = {'Pre Recorded Classes':5, 'Live Classes':8, 'Live Classes + Recorded':25}
Instruction_Preference = list(data.keys())
Number_of_respondants = list(data.values())
  
fig = plt.figure(figsize = (6, 5))
 
plt.barh(Instruction_Preference, Number_of_respondants, color ='r', alpha= 0.3)
 
plt.ylabel("Instruction preference", size=12)
plt.xlabel("Number of respondants", size =12)
plt.title("Instruction Preference of CIS 9655 students")
plt.xticks(range(0,35,5))
plt.annotate(xy = [25, 2], text = '65%', size=12)
plt.annotate(xy = [8, 1], text = '22%', size=12)
plt.annotate(xy = [5, 0], text = '13%', size=12)
fig.savefig("IP for CLass.png")
plt.show()
```

```python
data = {'In Person':27, 'Fully Remote':12, 'Hybrid':61} #IP=27 , FR=12 , H=61
Class_Preference = list(data.keys())
Number_of_respondants = list(data.values())
  
fig = plt.figure(figsize = (6, 5))
 
plt.barh(Class_Preference, Number_of_respondants, color ='m', alpha= 0.5)
 
plt.ylabel("Class preference", size=12)
plt.xlabel("Percentage of respondants", size =12)
plt.title("Class Preference for random sample of national students")
plt.xticks(range(0,105,10))
plt.annotate(xy = [61, 2], text = '61%', size=12)
plt.annotate(xy = [12, 1], text = '12%', size=12)
plt.annotate(xy = [27, 0], text = '27%', size=12)
plt.show()
fig.savefig("CP for RSN.png")
```
